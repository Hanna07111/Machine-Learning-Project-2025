```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
```

```python
# 경로 설정 및 데이터로드
TRAIN_CSV = "/content/drive/MyDrive/Coursework/25-2 Machine Learning/Project/DataPreprocessing_final/train_set_AB.csv"
TEST_CSV  = "/content/drive/MyDrive/Coursework/25-2 Machine Learning/Project/DataPreprocessing_final/test_set_AB.csv"

train = pd.read_csv(TRAIN_CSV)
test  = pd.read_csv(TEST_CSV)
```

### Closed World

```python
# train/test unmonitored data 제거
train_cw = train[train["label"] != 95].copy()
test_cw = test[test["label"]!= 95].copy()

#x,y 데이터 분리
train_cw_y = train_cw["label"].copy()
test_cw_y = test_cw["label"].copy()

train_cw_x = train_cw.drop(columns = ["label"]).copy()
test_cw_x = test_cw.drop(columns = ["label"]).copy()

# clipping
Q1, Q99 = train_cw_x.quantile(0.01), train_cw_x.quantile(0.99)
train_cw_clip = train_cw_x.clip(lower=Q1, upper=Q99, axis=1)
test_cw_clip  = test_cw_x.clip(lower=Q1, upper=Q99, axis=1)

# robust scaling
scaler = RobustScaler()
train_cw_sc = pd.DataFrame(scaler.fit_transform(train_cw_clip), columns=train_cw_clip.columns, index=train_cw_clip.index)
test_cw_sc  = pd.DataFrame(scaler.transform(test_cw_clip),     columns=test_cw_clip.columns,  index=test_cw_clip.index)

# 다시 label 합치기
train_cw = pd.concat([train_cw_sc, train_cw_y], axis=1)
test_cw = pd.concat([test_cw_sc, test_cw_y], axis=1)

# 데이터 출력
train_cw.head()
test_cw.head()
```

```python
# 데이터 저장
train_cw.to_csv("/content/drive/MyDrive/Coursework/25-2 Machine Learning/Project/DataPreprocessing_final/closedworld_train.csv", index=False)
test_cw.to_csv("/content/drive/MyDrive/Coursework/25-2 Machine Learning/Project/DataPreprocessing_final/closedworld_test.csv",  index=False)
```

### Open World

```python
#x,y 데이터 분리
train_ow_y = train["label"].copy()
test_ow_y = test["label"].copy()

train_ow_x = train.drop(columns = ["label"]).copy()
test_ow_x = test.drop(columns = ["label"]).copy()

# clipping
Q1, Q99 = train_ow_x.quantile(0.01), train_ow_x.quantile(0.99)
train_ow_clip = train_ow_x.clip(lower=Q1, upper=Q99, axis=1)
test_ow_clip  = test_ow_x.clip(lower=Q1, upper=Q99, axis=1)

# robust scaling
scaler = RobustScaler()
train_ow_sc = pd.DataFrame(scaler.fit_transform(train_ow_clip), columns=train_ow_clip.columns, index=train_ow_clip.index)
test_ow_sc  = pd.DataFrame(scaler.transform(test_ow_clip),     columns=test_ow_clip.columns,  index=test_ow_clip.index)

# 다시 label 합치기
train_ow = pd.concat([train_ow_sc, train_ow_y], axis=1)
test_ow = pd.concat([test_ow_sc, test_ow_y], axis=1)

# 데이터 출력
train_ow.head()
test_ow.head()
```

```python
# 데이터 저장
train_ow.to_csv("/content/drive/MyDrive/Coursework/25-2 Machine Learning/Project/DataPreprocessing_final/openworld_train.csv", index=False)
test_ow.to_csv("/content/drive/MyDrive/Coursework/25-2 Machine Learning/Project/DataPreprocessing_final/openworld_test.csv",  index=False)
```
